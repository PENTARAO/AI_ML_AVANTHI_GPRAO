```python
df = df.rename(columns={'A':'Words'})
```

Add character count of each word as another column in the dataframe

```python
df['Words'] = df['Words'].fillna('')
df['Charcount'] = df['Words'].apply(len)
df
```

Add a column called value which is the sum of the numerical positions of the alphabets in the word
(Ex: aah = 1 + 1 + 8 = 10)

```python
def summing(word):
  sum = 0
  word.lower()
  for char in word:
    position = ord(char) - 96
    sum += position
  return sum

df['Values'] = df['Words'].apply(summing)
df
```
