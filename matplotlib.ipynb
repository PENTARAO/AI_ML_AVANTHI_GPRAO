```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints=[0,6]
ypoints=[0,250]
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
xpoints=[0,1,3,5,7,6]
ypoints=[0,3,5,6,7,250]
plt.plot(xpoints,ypoints,'o')
plt.show()
```

```python
xpoints=np.array([1,2,6,8])
ypoints=([3,8,1,10])
plt.plot(xpoints,ypoints,)
plt.show()
```

```python
xpoints=np.array([1,2,2,4,4,6,6])
ypoints=np.array([1,4,2,3,5,1,4])
plt.plot(xpoints,ypoints,marker='o')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=x**2
plt.plot(x,y,'-.k')
plt.show()
```

```python
xpoints=np.array([1,2,4,6,8,9,10,12])
ypoints=np.array([1,2,1,4,3,5,0,6])
plt.plot(xpoints,ypoints,'o-g',ms=20,mec='#00008b',mfc='#FF33CA',lw='30')
```

```python
x1=np.array([2,4,4,6,6,10,10,12])
y1=np.array([1,2,1,5,1,2,6,2])
x2=np.array([1,8,27])
y2=np.array([1,2,3])
plt.plot(x1,y1)
plt.plot(x2,y2)
```

```python
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y)
plt.grid(linewidth=2,color='#800080')
```

```python
x=[1,2,3]
y=[1,2,3]
plt.plot(x,y)
plt.grid(linewidth=2,color='#800080')
plt.show()
```

```python
x1=[1,2,3]
y1=[1,2,3]
plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.title('first graph')

x2=[1,2,3]
y2=[1,2,3]
plt.subplot(2,2,4)
plt.plot(x2,y2)
plt.title('fourth graph')

x3=[3,2,1]
y3=[1,2,3]
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.title('third graph')
x4=[3,2,1]
y4=[1,2,3]
plt.subplot(2,2,2)
plt.plot(x4,y4)
plt.title('second graph')
plt.show()
```

```python
x1=np.array([2,4,4,6,6,10,10,12])
y1=np.array([1,2,1,5,1,2,6,2])
x2=np.array([1,8,27])
y2=np.array([1,2,3])
plt.scatter(x1,y1)
plt.scatter(x2,y2)
```
